```python
def set_random_seed(seed):
    torch.backends.cudnn.deterministic = True
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    np.random.seed(seed)
    random.seed(seed)
```

Вы можете придерживаться любой адекватной струкуры кода, но мы советуем воспользоваться следующими сигнатурами функций. Лучше всего, если вы проверите ваши предсказания ассертом: так вы убережете себя от разных косяков, например, что вектор предсказаний состоит из всего одного числа. В любом случае, внимательно следите за тем, для каких тензоров вы считаете метрику RMSE. При случайном или намеренном введении в заблуждение проверяющие очень сильно разозлятся.

```python
from torch.utils.data import Dataset, TensorDataset, DataLoader
from sklearn.preprocessing import StandardScaler
```

```python
set_random_seed(2112)
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
scaler1 = StandardScaler()
scaler1.fit(y_train.reshape(-1, 1))
y_train = scaler1.transform(y_train.reshape(-1, 1))
y_test = scaler1.transform(y_test.reshape(-1, 1))
```

```python
y_train = y_train.reshape(-1)
y_test = y_test.reshape(-1)
y_test.shape
```

```python
train_set = TensorDataset(torch.tensor(X_train),torch.tensor(y_train))
train_loader = DataLoader(train_set, batch_size=128,
    shuffle=False, 
    num_workers=1)

test_set =TensorDataset(torch.tensor(X_test),torch.tensor(y_test))
test_loader =  DataLoader(test_set, batch_size=128,
    shuffle=False, 
    num_workers=1)
```

```python
import torch.nn.functional as F
from tqdm.notebook import tqdm
from IPython.display import clear_output
import matplotlib.pyplot as plt
```

```python
model = nn.Sequential(nn.Linear(90, 128), nn.ReLU(),  nn.Linear(128, 10), nn.ReLU(), nn.Linear(10, 1))
optimizer = torch.optim.Adam(model.parameters(), lr= 0.0002)
criterion = nn.MSELoss()
```

```python
def train(model, optimizer, criterion, train_loader, test_loader, scaler = None):
    '''
    params:
        model - torch.nn.Module to be fitted
        optimizer - model optimizer
        criterion - loss function from torch.nn
        train_loader - torch.utils.data.Dataloader with train set
        test_loader - torch.utils.data.Dataloader with test set
                      (if you wish to validate during training)
    '''
    loss_arr = []
    for epoch in range(25):
      for x_tr, y_tr in tqdm(train_loader):    # берем батч из трейн лоадера
          y_pred = model(x_tr.float())     
          #print(y_pred, y_tr.float())                   # делаем предсказания
          loss =torch.sqrt(criterion(y_pred, y_tr.float()))
          #print(y_pred, y_tr.float())      # считаем лосс
          #print(loss)
          loss.backward()                                # считаем градиенты обратным проходом
          optimizer.step()                               # обновляем параметры сети
          optimizer.zero_grad()                        # обнуляем посчитанные градиенты параметров
      
      if epoch >=0:
          y_pred = test(model, criterion, test_loader)
          y_val = test_loader.dataset.tensors[1]
          if scaler is not None:
            y_pred = torch.tensor(scaler1.inverse_transform(y_pred.numpy().reshape(1, -1))[0])
            y_val = torch.tensor(scaler1.inverse_transform(y_val.numpy().reshape(1, -1))[0])
          print(y_pred, y_val)
          clear_output(True)
          plt.figure(figsize=(10,8))
          loss_arr.append(torch.sqrt(criterion(y_pred, y_val)))
          fig, ax = plt.subplots()
          ax.plot(loss_arr)
          plt.show()
          print(f"Epoch: {epoch}, loss: {loss_arr[-1]}")

def test(model, criterion, test_loader):
    '''
    params:
        model - torch.nn.Module to be evaluated on test set
        criterion - loss function from torch.nn
        test_loader - torch.utils.data.Dataloader with test set
    ----------
    returns:
        predicts - torch.tensor with shape (len(test_loader.dataset), ),
                   which contains predictions for test objects
    '''
    predicts = []
    with torch.no_grad():                          # на валидации запрещаем фреймворку считать градиенты по параметрам
        for x_val, y_val in tqdm(test_loader):  # берем батч из валидационного лоадера
            y_pred = model(x_val.float())           # делаем предсказания
            for y in y_pred:
              predicts.append(y)
    #predicts = torch.ones(len(test_loader.dataset))
    print(predicts)
    return torch.tensor(predicts)
```

```python
assert test(model, criterion, test_loader).shape[0] == y_test.shape[0]
```

```python
train(model, optimizer, criterion, train_loader, test_loader, scaler = scaler)
```

```python
pred = test(model, criterion, test_loader)
pred = torch.tensor(scaler1.inverse_transform(pred.numpy().reshape(1, -1))[0])
```

```python
ans = torch.tensor(scaler1.inverse_transform(y_test.reshape(1, -1))[0])
print("rmse = ", torch.sqrt(criterion(pred, ans)))
```
